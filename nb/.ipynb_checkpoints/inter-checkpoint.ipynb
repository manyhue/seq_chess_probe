```python
from lib.chess import chess_move_labels as le


class model_params:
    seq_len = 140
    lr = 0.0005
    vocab_size = len(le.classes_)


class params:
    num_workers = 24
    files_per_epoch = 24
    max_games_per_file = 250
    batch_size = 160

    max_epochs = 45
    lr = 0.0005

    flush_epoch_units = False
    loss_every = 20
    vocab_size = len(le.classes_)

    index = 9
```

```python
# load rwkv and gpt


from models.nanoRWKV import *
from models.nanoGPT import *

rwkv_model = RWKV(GPTConfig.create(model_params, bias=True))

rwkv_model.load_from_pth(
    "out/RWKV__seq_len=140__vocab_size=7797__n_layer=12__n_head=12__n_embd=768__dropout=0.0__bias=True__lr=0.005__weight_decay=0.01__epoch=89-135.pth"
)

rwkv_model.to(device)

index = 9

rwkv_layer = rwkv_model.transformer.h[params.index].tmix


gpt_model = GPT(GPTConfig.create(model_params, bias=True))

gpt_model.load_from_pth(
    "out/GPT__seq_len=140__vocab_size=7797__n_layer=12__n_head=12__n_embd=768__dropout=0.0__bias=True__lr=0.005__weight_decay=0.01__epoch=86-132.pth"
)
gpt_model.to(device)


gpt_layer = gpt_model.transformer.h[params.index].attn
```

```python
from data.pgnSquare import *

dt = PGNSquareData(PGNSquareDataConfig.create(params))
# dt.preview(samples=20)
```

```python
# define functions

from lib.metrics import *
from models.linear_probe import *


def train_probe(model, layer, prefix=""):
    probe = LinearProbe(model, layer, ProbeConfig(n_classes=dt.classes))

    probe.prefix_filename(model.filename().split("_")[0] + prefix)

    trainer = ProbeTrainer(
        TrainerConfig.create(
            params,
            epoch_end_callback=[
                dt.shuffle_files,
            ],
            load_previous=True,
        )
    )

    trainer.fit(probe, dt.loaders())

    trainer._save_model(prefix="GPT")

    return trainer


def evaluate(trainer):
    eval_ms = MetricsFrame(
        [
            from_te(
                ms.MulticlassConfusionMatrix,
                "ConfusionMatrix",
                num_classes=dt.classes,
            ),
            from_te(
                ms.MulticlassAccuracy,
                "Accuracy",
                num_classes=dt.classes,
            ),
            # initially 0.2
        ],
        flush_every=0,
        xlabel=None,
    )

    trainer.eval(
        pred=True,
        loss=True,
        mfs=eval_ms,
    )
    plot_2dheatmap(eval_ms.dict["ConfusionMatrix"][0])
    print(eval_ms.dict["Accuracy"])
```

```python
# rwkv
trainer = train_probe(rwkv_model, rwkv_layer)
```

```python
# rwkv

evaluate(trainer)
```

```python
# gpt

trainer = train_probe(gpt_model, gpt_layer)
```

```python
evaluate(trainer)
```
